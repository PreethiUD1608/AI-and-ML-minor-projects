```python
#Transfer learning
#input-Frozen layers- trainable layers-output layers
'''frozen layers=leayers whose weights are kept fixed and not updated during training
*Preserves general features learned from large datset
*no back propagation updates, it remain constant
*used when the new data set is small and similar to the pre trained dataset
*computation cost is lower since fewer parameters are trained
*ex in CNN:Early convolution layers that capture edges,textures and basic shapes'''

'''Trainable layers=layers whose heights rae updated during training
*adapt to task specific features of new dataset
*updated through the back propagation based on new data
*used when the the data is large and significantly different from original tasks
*Higher computation cost as more parameters needed
*ex in CNN:later fully connnected convolutional layers
or deeper convolutional layers or fully-tuned features'''
```

```python
import tensorflow as tf
from tensorflow.keras import models,layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import MobileNetV2

#Data augmentation-memory test for the Ai models
#purpose of data augmentation-it is a technique used to inc diversity of a dataset withour actually collecting new data
'''It works by applying various transformations to the existing data to create new, modified versions of data that helps
the model generalize better. In this article, we will learn more about data augmentation.'''
#below code is to how an image should be given to the model
data_augmentation = models.Sequential([layers.RandomFlip('horizontal'),layers.RandomRotation(0.1),])

# call Pre trained Base("The expert")
base_model=MobileNetV2(input_shape=(160,160,3),include_top=False,weights='imagenet')#pixel size and rgb#include top layer
#we have to keep it in the variable and flip the image
inputs=tf.keras.Input(shape=(160,160,3))#giving input
x=data_augmentation(inputs)

x=base_model(x,training=False)#or base_model.trainable=False
#training is false here bcz we are freezing the behaviour(training)(using freezing layers)
base_model.trainable=False

#create classification layers to create the final model
x=layers.GlobalAveragePooling2D()(x)
'''GlobalAveragePooling2D :it is a function which converts large feature maps into a small summary
it also reduces overfitting and give key imp features'''
#creating a drop out layer
x=layers.Dropout(0.2)(x)
#dropout layer helps to turn off learning of some neurons
#without dropout layer model memorises more which leads to overfitting of the model

#output layer(create how the output would be)
outputs=layers.Dense(1)(x)#it is giving one output

#final model creation
model=tf.keras.Model(inputs,outputs)
model.summary()
#finally it gives output by layer by layer
'''1.input_layer_5=(None, 160, 160, 3)#didn't learnt anything
2.sequential_3=(None, 160, 160, 3)#flipping of images #didn't learnt anything
3.mobilenetv2=(None, 5, 5, 1280) #learning:the model is learning here
analyzing the images by taking
#1280  means=it learnts 1280 features bcz model learnts from so many images
4global average pooling2d=(None, 1280) #didn't learnt anything
5.dropout=(None, 1280) #didn't learnt anything
6.dense #LEARNT AND GIVES A SINGLE OUTPUT
'''
```
